```python
import os
import numpy as np
import pandas as pd
import cv2 as cv
import matplotlib.pyplot as plt
import csv
import tensorflow as tf
```

```python
filepath = input("Dataset Path: ")
```

```python
os.chdir(filepath)
```

```python
CLASSNAMES = os.listdir()
```

```python
DATA = list()
for c in CLASSNAMES:
  f = os.path.join(filepath, c)
  label = CLASSNAMES.index(c)
  for img in os.listdir(f):
    path_to_img = os.path.join(f, img)
    image = cv.resize(cv.imread(path_to_img),(180, 180))
    DATA.append([image, label])
```

```python
X = []
y = []

for _, __ in DATA:
  X.append(_)
  y.append(__)

X = np.array(X)
y = np.array(y)
```

```python
X = X/255
```

```python
X.shape
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
model = Sequential()
model.add(Conv2D(64, (3,3), activation = "relu"))
model.add(MaxPooling2D((2,2)))


model.add(Conv2D(64, (3,3), activation = "relu"))
model.add(MaxPooling2D((2,2)))

model.add(Flatten())

model.add(Dense(128, input_shape = (180, 180, 3), activation = "relu"))

model.add(Dense(20, activation = "softmax"))
```

```python
model.compile(optimizer = "adam", loss = "sparse_categorical_crossentropy", metrics = ["accuracy"])
```

```python
model.fit(X, y, epochs = 5, validation_split = 0.1)
```

```python
model.save("CNNmodel.h5")
```
